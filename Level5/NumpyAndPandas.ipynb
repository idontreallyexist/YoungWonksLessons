```python
def salaryCatagory(salary):
    if salary<4000:
        return "Low"
    elif salary<9000:
        return "Medium"
    else:
        return "High"
df=pd.read_csv("employee-dataset.csv")
df2=pd.pivot_table(df,index=["Department"],values=["Age","Salary"],aggfunc={"Age":"sum","Salary":"sum"})
print(df2)
df2=pd.pivot_table(df,index=["Department"],columns=["Country","Gender"],values=["Age","Salary"],aggfunc={"Age":"sum","Salary":"sum"})
print(df2)
df["SalaryNew"]=df["Salary"].map(salaryCatagory)
print(df)
```
