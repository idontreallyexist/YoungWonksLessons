```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import cv2
import os
import time
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import yfinance as yf
from datetime import timedelta
from datetime import datetime
from sklearn.model_selection import train_test_split

df=pd.read_csv("csv_data/diabetes2.csv")

np.random.seed(0)
x=df[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]]
y=df["Outcome"]
model=LogisticRegression(max_iter=800)
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2)
model.fit(xTrain,yTrain)
y_predict=model.predict(xTest)

accuracy=metrics.accuracy_score(yTest,y_predict)
cfMatrix=metrics.confusion_matrix(yTest,y_predict)
print(accuracy)
print(cfMatrix)
```

```python
df=pd.read_csv("csv_data/Iris.csv")

np.random.seed(0)
x=df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y=df["Species"]
model=LogisticRegression(max_iter=10)
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2)
model.fit(xTrain,yTrain)
y_predict=model.predict(xTest)

accuracy=metrics.accuracy_score(yTest,y_predict)
cfMatrix=metrics.confusion_matrix(yTest,y_predict)
print(accuracy)
print(cfMatrix)

arr1=df["SepalLengthCm"].to_numpy()
arr2=df["SepalWidthCm"].to_numpy()
arr3=df["PetalWidthCm"].to_numpy()
arr4=df["Species"].to_numpy()

ax=plt.axes(projection='3d')
for i in range(0,149):
    if arr4[i]=="Iris-setosa":
        ax.scatter3D(arr1[i],arr2[i],arr3[i],color='red')
    elif arr4[i]=="Iris-versicolor":
        ax.scatter3D(arr1[i],arr2[i],arr3[i],color='blue')
    else:
        ax.scatter3D(arr1[i],arr2[i],arr3[i],color='green')
plt.title("3D Scatter")
ax.set_xlabel("x",fontweight='bold')
ax.set_ylabel("y",fontweight='bold')
ax.set_zlabel("z",fontweight='bold')
```
