```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import yfinance as yf
from datetime import timedelta
from sklearn.model_selection import train_test_split

x=np.linspace(1,10,50).reshape(-1,1)
y=np.linspace(1,10,50)+np.random.random(50)

plt.scatter(x,y,alpha=0.7)

model=LinearRegression()
model.fit(x,y)
y_predict = model.predict(x)

plt.plot(x,y_predict,color="red",linewidth=1)

value=model.predict([[20]])
print(value)
plt.scatter(20,value,color="red",alpha=0.7)
plt.show()

print("m:",model.coef_[0],"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(y,y_predict))
print("R2 Score:",r2_score(y,y_predict))

```

```python
df=pd.read_csv("csv_data/population_total_long.csv")
world=df.groupby(["Year"])["Count"].sum()
world=world.reset_index()
x=world["Year"].to_numpy()
x=np.reshape(x,(-1,1))
model=LinearRegression()
model.fit(x,world["Count"])
y_predict = model.predict(x)
plt.scatter(x,world["Count"],alpha=0.7)
plt.plot(x,y_predict,color="red",linewidth=1)
y=world["Count"]

print("m:",model.coef_[0],"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(y,y_predict))
print("R2 Score:",r2_score(y,y_predict))

value=model.predict([[2020]])
value2=model.predict([[2050]])

plt.scatter(2020,value,color="red",alpha=0.7)
plt.scatter(2050,value2,color="red",alpha=0.7)
plt.show()
```

```python
df=pd.read_csv("csv_data/population_total_long.csv")
fig,subplots=plt.subplots(1,3)
world=df[df["Country Name"]=="United States"].groupby(["Year"])["Count"].sum()
world=world.reset_index()
x=world["Year"].to_numpy()
x=np.reshape(x,(-1,1))
model=LinearRegression()
model.fit(x,world["Count"])
y_predict = model.predict(x)
y=world["Count"]

print("m:",model.coef_[0],"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(y,y_predict))
print("R2 Score:",r2_score(y,y_predict))

x2=["2017","2050"]
value=model.predict([[2050]])
y=[world[world["Year"]==2017]["Count"].values[0],value[0]]

subplots[0].pie(y,labels=x2,autopct="%1.1f%%")
plt.show()
```

```python
fig,subplots=plt.subplots(1,3)
df=pd.read_csv("csv_data/Advertising.csv")
x=df[["TV","radio","newspaper"]]
y=df["sales"].to_numpy()
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(xTrain,yTrain)
y_predict = model.predict(xTrain)
y_predict2 = model.predict(xTest)
xBar=[x for x in range(1,201)]
yBar=[]

subplots[0].scatter(xTest["TV"],y_predict2,color='blue',alpha=0.7)
subplots[0].scatter(xTest["TV"],yTest,color='red',alpha=0.7)
subplots[1].scatter(xTest["radio"],y_predict2,color='blue',alpha=0.7)
subplots[1].scatter(xTest["radio"],yTest,color='red',alpha=0.7)
subplots[2].scatter(xTest["newspaper"],y_predict2,color='blue',alpha=0.7)
subplots[2].scatter(xTest["newspaper"],yTest,color='red',alpha=0.7)
plt.show()
print("m:",model.coef_,"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(yTest,y_predict2))
print("R2 Score:",r2_score(yTest,y_predict2))
```

```python
df=yf.download("AAPL",start='2019-1-1',end='2019-12-31').reset_index()
x=df["Date"]
y=df["Close"]
plt.plot(x,y,alpha=0.7)
x2=df.index.to_numpy()
x2=np.reshape(x2,(-1,1))
model=LinearRegression()
model.fit(x2,y)
y_predict = model.predict(x2)
plt.plot(x,y_predict,color="red",linewidth=1)
new=df.tail(1)
new["Date"]=pd.to_datetime(new["Date"])
print(new["Date"])
for i in range(252,300):
    value=model.predict([[i]])
    new["Date"]+=timedelta(days=1)
    plt.scatter(new["Date"],value,color="red",alpha=0.7)
plt.show()

print("m:",model.coef_[0],"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(y,y_predict))
print("R2 Score:",r2_score(y,y_predict))
```

```python
from sklearn.datasets import load_diabetes
fig,subplots=plt.subplots(1,3)
df=load_diabetes(as_frame=True)["frame"]
x=df[["age","sex","bmi"]]
y=df["target"].to_numpy()
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(xTrain,yTrain)
y_predict = model.predict(xTrain)
y_predict2 = model.predict(xTest)

subplots[0].scatter(xTest["age"],y_predict2,color='blue',alpha=0.7)
subplots[0].scatter(xTest["age"],yTest,color='red',alpha=0.7)
subplots[1].scatter(xTest["sex"],y_predict2,color='blue',alpha=0.7)
subplots[1].scatter(xTest["sex"],yTest,color='red',alpha=0.7)
subplots[2].scatter(xTest["bmi"],y_predict2,color='blue',alpha=0.7)
subplots[2].scatter(xTest["bmi"],yTest,color='red',alpha=0.7)
plt.show()
print("m:",model.coef_,"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(yTest,y_predict2))
print("R2 Score:",r2_score(yTest,y_predict2))
```

```python
fig,subplots=plt.subplots(1,3)
df=pd.read_csv("csv_data/auto-mpg.csv")
x=df[["horsepower","cylinders","weight"]]
x.replace(to_replace='?',value="0",inplace=True)
y=df["mpg"].to_numpy()
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(xTrain,yTrain)
y_predict = model.predict(xTrain)
y_predict2 = model.predict(xTest)

subplots[0].scatter(xTest["horsepower"],y_predict2,color='blue',alpha=0.7)
subplots[0].scatter(xTest["horsepower"],yTest,color='red',alpha=0.7)
subplots[1].scatter(xTest["cylinders"],y_predict2,color='blue',alpha=0.7)
subplots[1].scatter(xTest["cylinders"],yTest,color='red',alpha=0.7)
subplots[2].scatter(xTest["weight"],y_predict2,color='blue',alpha=0.7)
subplots[2].scatter(xTest["weight"],yTest,color='red',alpha=0.7)
plt.show()
print("m:",model.coef_,"b:",model.intercept_)
print("Mean Square Error:",mean_squared_error(yTest,y_predict2))
print("R2 Score:",r2_score(yTest,y_predict2))
```
