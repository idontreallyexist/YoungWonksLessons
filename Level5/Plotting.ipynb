```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import random
from datetime import datetime

x=[2,1,3,6,8]
y=[1,8,5,6,3]
#plt.scatter(x,y)
#plt.plot(x,y)

fig=plt.figure(figsize=(10,5))
plt.bar(x,y,color='orange',width=0.5)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot",fontsize=12)

plt.show()
```

```python
x=["Apples","Oranges","Bananas","Grapes","Watermelons"]
y=[2,1,3,6,8]

plt.pie(y,labels=x,autopct="%1.1f%%")

plt.xlabel("X")
plt.ylabel("Fruits")
plt.title("Plot",fontsize=12)
plt.legend(fontsize=12)

plt.show()
```

```python
x=["Apples","Oranges","Bananas","Grapes","Watermelons"]
y=[2,1,3,6,8]

fig,subplots=plt.subplots(2,2)

#plt.subplot(1,4,1,label="Line Plot")
subplots[0,0].plot(x,y)
#plt.subplot(1,4,2,label="Scatter Plot")
subplots[0,1].scatter(x,y)
#plt.subplot(1,4,3,label="Bar Graph")
subplots[1,0].bar(x,y)
#plt.subplot(1,4,4,label="Pie Chart")
subplots[1,1].pie(y,labels=x,autopct="%1.1f%%")

#fig.autofmt_xdate(rotation=45)
plt.suptitle("Ploting Types")
subplots[0,0].tick_params(axis="x",rotation=45)
subplots[0,1].tick_params(axis="x",rotation=45)
subplots[1,0].tick_params(axis="x",rotation=45)
subplots[1,1].tick_params(axis="x",rotation=45)
fig.tight_layout()
plt.show()
```

```python
arr1=[25,63,51,78,44]
arr2=[8,7,5,9,4]
nums=["num1","num2","num3","num4","num5"]
width=0.35
index=[0,1,2,3,4]
fig=plt.figure(figsize=(10,7))
p1=plt.bar(index,arr1,width)
p2=plt.bar(index,arr2,width,bottom=arr1)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Stacked Bars")

plt.show()
```

```python
barwidth=0.25
fig=plt.subplots(figsize=(12,8))

arr1=[19,17,12,26,43]
arr2=[27,43,10,7,26]
arr3=[1,8,6,40,19]

index1=[0,1,2,3,4]
index2=[]
for x in index1:
    index2.append(x+barwidth)
index3=[x+barwidth for x in index2]

plt.bar(index1,arr1,color='gold',width=barwidth,edgecolor='black',label='Gold')
plt.bar(index2,arr2,color='grey',width=barwidth,edgecolor='black',label='Silver')
plt.bar(index3,arr3,color='brown',width=barwidth,edgecolor='black',label='Bronze')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Multibar Graph")
plt.show()
```

```python
import urllib


imagelist=[]
url='https://upload.wikimedia.org/wikipedia/commons/thumb/9/97/The_Earth_seen_from_Apollo_17.jpg/500px-The_Earth_seen_from_Apollo_17.jpg'
imagefile=img.imread(urllib.request.urlopen(url))
imagelist.append(imagefile)

plt.figure()
plt.imshow(imagelist)
plt.show()
```

```python
arr1=[random.randint(0,10) for _ in range(0,20)]
arr2=[random.randint(0,10) for _ in range(0,20)]
arr3=[random.randint(0,10) for _ in range(0,20)]

ax=plt.axes(projection='3d')
for x in range(0,20):
    ax.scatter3D(arr1[x],arr2[x],arr3[x],color='red',marker='^')
plt.title("3D Scatter")
ax.set_xlabel("x",fontweight='bold')
ax.set_ylabel("y",fontweight='bold')
ax.set_zlabel("z",fontweight='bold')

plt.show()
```

```python
arr1=[random.randint(0,10) for _ in range(0,20)]
arr2=[random.randint(0,10) for _ in range(0,20)]
arr3=[0 for _ in range(0,20)]

d1=[1 for _ in range(0,20)]
d2=[1 for _ in range(0,20)]
d3=[random.randint(0,10) for _ in range(0,20)]

ax=plt.axes(projection='3d')
ax.bar3d(arr1,arr2,arr3,d1,d2,d3,color='red')

plt.title("3D Bar")
ax.set_xlabel("x",fontweight='bold')
ax.set_ylabel("y",fontweight='bold')
ax.set_zlabel("z",fontweight='bold')

plt.show()
```

```python
dates=[datetime(2019,random.randint(1,12),random.randint(0,27)) for _ in range(0,10)]
y=10

for x in dates:
    plt.plot_date(x,y)

plt.gcf().autofmt_xdate()

plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Datetime plots")
plt.tight_layout()
plt.show()
```
